```python
# import os os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
import numpy as np
import pandas as pd
gpd.options.use_pygeos = True
```

```python
# read results
gdf_results = gpd.read_file('./voronoi_data/').to_crs(3857)
```

```python
gdf_results.head()
```

```python
gdf_results['VB_area_di'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the x-axis limit
plt.xlim(0, 0.5*1e5)

sns.histplot(gdf_results['VB_area_di'], bins=10000, label='VB Area Difference', kde=True)

# Show the legend
plt.legend()

# Show the plot
plt.savefig('./plots/area_difference_plot.png')
plt.show()
```
