```python
import re
import matplotlib.pyplot as plt
import os
```

```python
# Feed in log files
filenames = ['loss_log_nc-charlotte-200-150-v100_344-epochs.txt', 'loss_log-ma-boston-p2p-200-150-v100_362-epochs.txt', 'loss_log-ny-manhattan-p2p-200-150-v100_306-epochs.txt']
```

```python
# Initialize lists to hold data for all files combined
all_epochs = []
all_iters = []
all_G_GAN = []
all_G_L1 = []
all_D_real = []
all_D_fake = []

# Read and parse each log file
for i, filename in enumerate(filenames):
    # Initialize lists to hold data for this file
    epochs = []
    iters = []
    G_GAN = []
    G_L1 = []
    D_real = []
    D_fake = []

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                epochs.append(int(match.group(1)))
                iters.append(int(match.group(2))+int(match.group(1))*20100-20100)
                G_GAN.append(float(match.group(3)))
                G_L1.append(float(match.group(4)))
                D_real.append(float(match.group(5)))
                D_fake.append(float(match.group(6)))

# Get the filename without the extension
filename_without_ext = os.path.splitext(filename)[0]

# Plot data for this file
plt.figure()
plt.plot(epochs, G_GAN)
plt.title(f'G_GAN for {filename}')
plt.savefig(f'graph-{filename_without_ext}.png')
# plt.show(){filename}.png')
plt.show()



#plt.tight_layout()
#plt.savefig(f'graph-{filename_without_ext}.png')
#plt.show()
```
