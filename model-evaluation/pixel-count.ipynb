```python
import cv2
import numpy as np
import os, sys
```

```python
mask2 = np.load("segmentation_mask.npy")
```

```python
dict_class_area = {}
for class_num in np.unique(mask2):
    
    pimg = np.zeros(mask2.shape)
    mask_class = (mask2 == class_num)
    pimg[mask_class] = 255
    
    retval, labels, stats, centroids = cv2.connectedComponentsWithStats(pimg.astype(np.uint8))
    total_class_area = []
    
    for i, st in enumerate(stats):
        imask = (labels == i)
        vave = pimg[imask].mean()
        if vave < 128:  #　１２８
            continue
        area = st[cv2.CC_STAT_AREA]
        total_class_area.append(area)
    dict_class_area[class_num] = sum(total_class_area)
```
